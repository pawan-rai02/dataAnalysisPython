```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset= load_dataset('lukebarousse/data_jobs')
df= dataset['train'].to_pandas()
```

```python
'''Get the first 20 rows and first 5 columns using .iloc'''
df.iloc[:20, :5]
```

```python
'''Get the job_title_short and job_location of rows 100 to 110 using .loc.'''
df.loc[100:110, ['job_title_short', 'job_location']]
```

```python
'''Get the company_name of the last row'''
df.loc[:-1, 'company_name']
```

```python
'''Using .loc, filter jobs where job_country == "India" and display only company_name and job_title_short.'''

df.loc[df['job_country'] == 'India', ['job_title_short', 'company_name']]
```
