```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset= load_dataset('lukebarousse/data_jobs')
df= dataset['train'].to_pandas()
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])


df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')
df['job_posted_month'].head()
```

```python
months = df['job_posted_month'].unique()
months
```

```python
dict_months = {month : df[df['job_posted_month'] == month] for month in months}
dict_months['Feb']
```

```python
df_q2 = pd.concat([dict_months['Apr'], dict_months['May'], dict_months['Jun']])
df_q2
```

```python
df_q2.value_counts('job_title_short').head(10).plot(kind='barh', figsize=(10, 6))
```

```python
df_q2['job_posted_month'].value_counts().plot(kind = 'bar')
```

```python
df_q4 = pd.concat([dict_months['Oct'], dict_months['Nov'], dict_months['Dec']] )
df_q4['job_posted_month'].value_counts().plot(kind = 'bar') 
plt.show()
```
