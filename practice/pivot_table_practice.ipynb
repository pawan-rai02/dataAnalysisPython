```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.head()
```

```python
df.pivot_table(index='job_title_short', values='salary_year_avg', aggfunc='median')
```

```python
df.pivot_table(index='job_title_short', columns='job_country', values='salary_year_avg', aggfunc='median')       
```

```python
df.pivot_table(columns= 'job_country', index='company_name', values= 'job_title_short', aggfunc= 'count').dropna()
```

```python
df.pivot_table(index= 'company_name', values= 'salary_year_avg', aggfunc= 'max').sort_values(by= 'salary_year_avg')
```

```python
df.pivot_table(index= 'job_title_short', values= 'salary_year_avg', aggfunc= ['max', 'min', 'median'])
```
