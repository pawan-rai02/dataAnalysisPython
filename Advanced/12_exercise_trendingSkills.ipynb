```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import ast

dataset= load_dataset('lukebarousse/data_jobs')
df= dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst']
df_DA.head(10)
```

```python
df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month
df_DA.head()
```

```python
df_DA_explode = df_DA.explode('job_skills')
df_DA_explode
```

```python
df_DA_pivot = df_DA_explode.pivot_table(index= 'job_posted_month_no', columns= 'job_skills', aggfunc= 'size', fill_value=0)
```

```python
df_DA_pivot.loc['Total']  = df_DA_pivot.sum()
df_DA_pivot
```

```python
df_final = df_DA_pivot.loc['Total'].sort_values(ascending= False).head()
df_final.plot(kind= 'barh')

                                     
```

```python
df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending= False).index]
df_DA_pivot = df_DA_pivot.drop('Total')
df_DA_pivot
                
```

```python
df_DA_pivot.iloc[:, :5].plot(kind= 'line')
```

```python
df_DS = df[df['job_title_short'] == 'Data Engineer']
df_DS
```

```python
df_DS = df_DS.explode('job_skills')
df_DS
```

```python
df_DS['job_posted_month'] = df_DS['job_posted_date'].dt.strftime('%b')
df_DS['job_posted_month']
```

```python
df_DS_pivot = df_DS.pivot_table(index= 'job_posted_month', columns= 'job_skills', aggfunc= 'size', fill_value= 0)
df_DS_pivot['job_posted_month'].dt.month
```

```python
df_DS_pivot.loc['total'] = df_DS_pivot.sum()
df_DS_pivot
```

```python
df_DS_pivot = df_DS_pivot[df_DS_pivot.loc['total'].sort_values(ascending= False).index]
```

```python
df_DS_pivot = df_DS_pivot.drop('total')
df_DS_pivot
```

```python
df_DS['month no'] = df_DS['job_posted_date'].dt.month
df_DS = df_DS.sort_values(by='month no')
df_DS.reset_index()
```
